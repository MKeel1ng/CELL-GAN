##AI generated tattoos

Hello. In this small site I'd like to showcase some AI generated tattoos and the cool capabilities that this process let's us play with!

Pick a seed to generate an image. Adjust the truncation to change the quality. Lower truncation means better quality, but lower variety. Higher truncation means more different images, but with lower quality.



```
seedSlide=widgets.IntSlider(
    value=1,
    min=1,
    max=100,
    step=1,
    description='Seed:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

truncSlide=widgets.FloatSlider(
    value=0.5,
    min=0.0,
    max=1.0,
    step=0.25,
    description='Truncation:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)

out_pl1 = widgets.Output()


btn_run=widgets.Button(
    description='Show',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click me',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

def on_click_Generate(change):
  


  truncVal=truncSlide.value
  if truncVal==0.0:
    path=path0
  elif truncVal==0.25:
   path=path025
  elif truncVal==0.75:
   path=path075
  elif truncVal==1.0:
   path=path01
  else:
    path=path050


  im = Image.open(path+f'seed{seedSlide.value:04d}.png')

  out_pl1.clear_output()
  with out_pl1:
      display(im)

btn_run.on_click(on_click_Generate)


widgets.VBox([seedSlide,truncSlide,btn_run,out_pl1])
```
